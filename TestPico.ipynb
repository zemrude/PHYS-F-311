```python
import pandas as pd
```

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
```

```python
columns = ["Time", "Q"]
df = pd.read_csv("Integral.csv", names=columns, skiprows=1)
```

```python
print(df)
```

```python
data1 = -1 * df["Q"].values
```

```python
fig, ax = plt.subplots(1,1, figsize=(15, 9))
ax.hist(data1, bins=np.linspace(0, 6e-8, 100))
ax.set_xlim(0, 0.6e-7)
ax.set_xlabel("V*s")
plt.show()
```
